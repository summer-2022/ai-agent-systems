### Library

```python
import os
from google.colab import userdata
from openai import OpenAI
import requests
import json
```

### API Function

```python
BASE_URL = "https://nomad-movies.nomadcoders.workers.dev"

def get_popular_movies():
    response = requests.get(f"{BASE_URL}/movies")
    return response.json()

def get_movie_details(id):
    response = requests.get(f"{BASE_URL}/movies/{id}")
    return response.json()

def get_movie_credits(id):
    response = requests.get(f"{BASE_URL}/movies/{id}/credits")
    return response.json()
```

###  OpenAI Client Setup

```python
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")

client = OpenAI()
```

### Tool Definitions for the Agent

```python

tools = [
    {
        "type": "function",
        "function": {
            "name": "get_popular_movies",
            "description": "Get a list of popular movies",
            "parameters": {
                "type": "object",
                "properties": {}
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_movie_details",
            "description": "Get details of a movie by ID",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {
                        "type": "integer",
                        "description": "Movie ID"
                    }
                },
                "required": ["id"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "get_movie_credits",
            "description": "Get cast and crew of a movie by ID",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {
                        "type": "integer",
                        "description": "Movie ID"
                    }
                },
                "required": ["id"]
            }
        }
    }
]
```

### 5. Test 1: Get Popular Movies

```python
user_input = "지금 인기 있는 영화가 무엇인지 알려줘"

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a movie expert agent."},
        {"role": "user", "content": user_input}
    ],
    tools=tools,
    tool_choice="auto"
)

response
```

```python
tool_call = response.choices[0].message.tool_calls[0]

function_name = tool_call.function.name
arguments = json.loads(tool_call.function.arguments)

print(function_name, arguments)

if function_name == "get_popular_movies":
    result = get_popular_movies()

elif function_name == "get_movie_details":
    result = get_movie_details(**arguments)

elif function_name == "get_movie_credits":
    result = get_movie_credits(**arguments)

print(result)
```
